```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
import pandas as pd
```

```python
df = pd.read_pickle('tweets_preprocessed.pkl')
df.head(3)
```

1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.

Игнорируем слова, частота которых в документе строго превышает порог 0.9 с помощью max_df.
Ограничим количество слов, попадающий в мешок, с помощью max_features = 1000.
Исключим стоп-слова с помощью stop_words='english'. 
Отобразим Bag-of-Words модель как DataFrame. columns необходимо извлечь с помощью CountVectorizer.get_feature_names().


```python
count_vectorizer = CountVectorizer(stop_words='english', max_df=0.9, max_features=1000)
```

```python
stemmed_corpus = list(set(df.tweet_stemmed.sum()))
```

```python
bag_of_words_cv_stemmed = count_vectorizer.fit_transform(stemmed_corpus)
```

```python
feature_names = count_vectorizer.get_feature_names()
```

```python
feature_names
```

```python
# With limit of 1000 we getting rather silly features IMO... )) Only words starting with P
```

```python
pd.DataFrame(bag_of_words_cv_stemmed.toarray(), columns = feature_names).head()
```

```python
lemmatized_corpus = list(set(df.tweet_lemmatized.sum()))
```

```python
len(lemmatized_corpus)
```

```python
bag_of_words_cv_lemmatized = count_vectorizer.fit_transform(lemmatized_corpus)
```

```python
feature_names = count_vectorizer.get_feature_names()
```

```python
feature_names
```

```python
# Same shit here ) only P words. Think we should increase max_features
```

```python
pd.DataFrame(bag_of_words_cv_lemmatized.toarray(), columns = feature_names).head()
```
